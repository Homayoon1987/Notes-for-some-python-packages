# Data Visualization

Import necessary packages:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

To show plots in the same window:

```python
%matplotlib inline 
```

Lets create some points:

```python
X = np.arange(0,10)
y = X**2
```

Let's plot y vs X

```python
plt.plot(X,y)
```

```python
plt.plot(X,y,'*')
```

```python
plt.plot(X,y,'red')
```

```python
plt.plot(X,y,'r--')
```

How to edit axis and title?

```python
plt.plot(X,y,'r--')
plt.xlim(2,4)
plt.ylim(10,20)
plt.title("Zoomed")
plt.xlabel("X axis")
plt.ylabel("y axis")
```

Can we plot matrix?


```python
matrix = np.arange(0,100).reshape(10,10)
plt.imshow(matrix)
```

```python
matrix = np.random.randint(0,100,(10,10))
plt.imshow(matrix,cmap='coolwarm')
plt.colorbar()
```
